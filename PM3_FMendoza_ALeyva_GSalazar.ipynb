La solución a este sistema `masa-resorte` se explica en términos de la segunda ley de Newton. Para este caso, si la masa permanece constante y solo consideramos la dirección en $x$. Entonces,
\begin{equation}
F = m \frac{d^2x}{dt^2}.
\end{equation}

¿Cuál es la fuerza? **Ley de Hooke!**
\begin{equation}
F = -k x, \quad k > 0.
\end{equation}

Vemos que la fuerza se opone al desplazamiento y su intensidad es proporcional al mismo.  Y $k$ es la constante elástica o recuperadora del resorte. 

Entonces, un modelo del sistema `masa-resorte` está descrito por la siguiente **ecuación diferencial**:

\begin{equation}
 M\frac{d^2x}{dt^2} + B\frac{dx}{dt}+ \frac{k}{m}x = f(t),
\end{equation}

cuya solución se escribe como 
\begin{equation}
\frac{dx}{dt} = v
\end{equation}
Y su primera derivada (velocidad) sería 
\begin{equation}
\frac{dv}{dt} = \frac{ft}{M} -B\frac{dx}{dt} -\frac{k}{m}x
\end{equation}

+ Donde M es la masa oscilante
+ Donde B es la constante de fricción 
+ Donde K es la constante de elasticidad del resorte
+ Donde f(x) es la fuerza externa al sistema

**Objetivos Específicos**
Modelar un Sistema que entre en resonancia debido a la aplicacion de una fuerza externa f(t) 
Identificar el comportamiento de un Sistema en resonancia mediante el uso de gráficas.


```python
import numpy as np
import sympy as sym
sym.init_printing(use_latex='mathjax')
import matplotlib.pyplot as plt
from scipy.integrate import odeint
    
```

```python
def oscilador(x,t):
    x1=x[0]
    x2=x[1]
    b=-35.6
    m=99
    k=(5*np.pi)**2*m
   
    ft=2*np.cos(0.01*np.pi*t)
    dx1= x2
    dx2= ft/m - (b/m)*x2- (k/m)*x1
    return np.array([dx1,dx2])
```

```python
a,b=0,0

x0=np.array([a,b])
t=np.linspace(0,5,1000)



dx=odeint(oscilador,x0,t)
x1=dx[:,0]
x2=dx[:,1]
   
```

```python
plt.figure(figsize=(15,15))
plt.plot(t, x1, label='$ɵ(t)$')
#plt.plot(t, x2, label='$w(t)$')
plt.xlabel('Tiempo $t$')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.plot(x1,x2,label="$0(t)$")
plt.xlabel("Tiempo $t$")
plt.axhline(y=0, color="k")
plt.axvline(x=0, color="k")
plt.legend(loc="best")
plt.grid()
plt.show()
```
